```python
with open('movie即將上映20部.csv',encoding='utf-8',newline='') as movie:
    csvread=csv.reader(movie)
    movie_data=list(csvread)
movie_data=movie_data[1:]
movie_data
```

```python
text=[]
#print(movie_data[:1])
for i in range(10,20):
    name=movie_data[i][0]
    date=movie_data[i][2]
    img_url=movie_data[i][3]
    movie_url=movie_data[i][4]
    content=name
    bubble={
     'type': 'bubble',
     'hero': {'type': 'image',
      'margin': 'none',
      'size': 'full',
      'url': img_url,
      'aspectMode': 'fit'},
        
     'body': {'type': 'box',
              'layout': 'vertical',
              'contents': [{'type': 'text', 'text': name, 'weight': 'bold', 'size': 'xxl'},
                           {'type': 'text', 'text': '上映日期：'+str(date)},
                           {'type': 'box',
                            'layout': 'horizontal',
                            'contents': [{'type': 'button',
                                          'action': {'type': 'message',
                                                     'label': '電影簡介',
                                                     'text': content}},
                                         {'type': 'button',
                                          'action': {'type': 'uri',
                                           'label': '預告片',
                                           'uri': movie_url}

         }]}]}}
    text.append(bubble)
message={'type': 'carousel',
 'contents':text}
message
```

```python
with open("bubble_page2.json",'w',encoding='utf-8') as f:
    json.dump(message,f)
```
